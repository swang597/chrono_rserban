```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('viper_wheel_obj.txt')
# data = np.loadtxt('viper_cylwheel_obj.txt')[1::2,:]
```

```python
# plot 3D data
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:,0], data[:,1], data[:,2], c='r', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()



```

```python
z_min, z_max = np.min(data,axis=0), np.max(data,axis=0)

for i in range(3):
    print(z_min[i], z_max[i], (z_max[i]-z_min[i])*0.5)
```

```python
x_min, x_max = np.min(data[:,0]), np.max(data[:,0])
print(x_min, x_max, x_max-x_min)
```
